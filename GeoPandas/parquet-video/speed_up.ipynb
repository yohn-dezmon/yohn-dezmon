```python
import geopandas
```

```python
# without using parquet, read_file() takes ~6 min
# the geojson file is 617 MB
nyc_trees = geopandas.read_file("data/2015 Street Tree Census - Tree Data.geojson")
```

```python
# takes < 1 min with parquet
# note: you need to install pyarrow for this to work
nyc_trees = geopandas.read_parquet("data/nyc_trees.parquet")
```

### Warning: 
* The GeoPandas documentation states that to_parquet is not reliable for production unless you are ok with rewriting the parquet file after updates
* The reason being that they are using an early version of GeoParquet and future updates could introduce breaking changes

### WKB
* Any geometry columns present are serialized to WKB format in the file.
* Well-Known Binary format
* each geometry object is encoded as a series of bytes
* only supports simple geometries like points, lines and polygons
* another related encoding is WKT (Well-Known Text)

### Parquet
* apache file format
* column oriented 
* allows for better compression, and thus faster reads

```python
nyc_trees.to_parquet('data/nyc_trees.parquet')
```
