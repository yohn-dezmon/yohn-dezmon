```python
nybb = geopandas.read_file("data/nybb.geojson")
# simplify geometry to save sapce wehn rendering many interactive maps
nybb.geometry = nybb.simplify(200)
```

```python
nybb.head()
```

```python
nybb.explore()
```

```python
n200_sampled_points =  nybb.sample_points(200)
m = nybb.explore()
# m stands for map, it is the base layer... 
n200_sampled_points.explore(m=m, color='red')
```

```python
manhattan_parts = nybb.iloc[[3]].explode(ignore_index=True)
manhattan_parts.head()
```

```python
# here we get the 3rd row, I guess bc that is the row that has manhattan data
# explode() transforms a column containing lists into multiple rows!
# in this case, manhattan's geometry is a MultiPolygon!
manhattan_parts = nybb.iloc[[3]].explode(ignore_index=True)
# not sure why we get the 30th row...
manhattan_island = manhattan_parts.iloc[[30]]
manhattan_island.boundary.explore()
```

```python
manhattan_border_points = manhattan_island.boundary.sample_points(200)
m = manhattan_island.explore()
manhattan_border_points.explore(m=m, color='red')
```

```python
manhattan_border_points
```

* the sampled points are returned as a single multi-part geometry
* the distances over the line segments are calculated along the line

```python
# you can separate out the individual sampled points 
# with .explode()
manhattan_border_points.explode(ignore_index=True).head()
```

## Variable number of points
* you can sample a different number of points from different geometries if you pass an array specifying hte size of the sample per geometry

```python
variable_size = nybb.sample_points([10, 50, 100, 200, 500])
m = nybb.explore()
variable_size.explore(m=m, color='red')
```

## Sampling from more complicated point pattern processes

* the `sample_points()` method can use different sampling processes than those described above. 

A "cluster-poisson" process is a spatially random cluster process where the "seeds" of clusters are chosen randomly, and then points around these clusters are distributed randomly as well. 

* here we will have 10 points distributed around 5 seeds within each of the boroughs of NYC
