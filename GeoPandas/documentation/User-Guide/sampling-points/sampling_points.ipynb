```python
import geopandas
import geodatasets
```

```python
nybb = geopandas.read_file("data/nybb.geojson")
# simplify geometry to save sapce wehn rendering many interactive maps
nybb.geometry = nybb.simplify(200)
```

```python
nybb.head()
```

```python
nybb.explore()
```

```python
n200_sampled_points =  nybb.sample_points(200)
m = nybb.explore()
# m stands for map, it is the base layer... 
n200_sampled_points.explore(m=m, color='red')
```

```python
manhattan_parts = nybb.iloc[[3]].explode(ignore_index=True)
manhattan_parts.head()
```

```python
# here we get the 3rd row, I guess bc that is the row that has manhattan data
# explode() transforms a column containing lists into multiple rows!
# in this case, manhattan's geometry is a MultiPolygon!
manhattan_parts = nybb.iloc[[3]].explode(ignore_index=True)
# not sure why we get the 30th row...
manhattan_island = manhattan_parts.iloc[[30]]
manhattan_island.boundary.explore()
```

```python
manhattan_border_points = manhattan_island.boundary.sample_points(200)
m = manhattan_island.explore()
manhattan_border_points.explore(m=m, color='red')
```
